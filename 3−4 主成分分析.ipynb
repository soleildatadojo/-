```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.random(size=70)
Y = 2*X + 0.5*np.random.rand(70)

fig, ax = plt.subplots()
ax.scatter(X, Y)
plt.show()
```

```python
from sklearn.decomposition import PCA

# インスタンスを生成
pca = PCA(n_components=2)

# 分析を実行
X_pca = pca.fit_transform(np.hstack((X[:, np.newaxis], Y[:, np.newaxis])))
```

```python
np.hstack((X[:, np.newaxis], Y[:, np.newaxis])).shape
```

```python
fig, ax = plt.subplots()
ax.scatter(X_pca[:, 0], X_pca[:, 1])
ax.set_xlabel('variable1')
ax.set_ylabel('variable2')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
plt.title('PCA')
plt.show()
```
