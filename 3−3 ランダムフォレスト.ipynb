# ▶︎ 決定木分析

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# データをインポート
iris = load_iris()
X = iris.data
y = iris.target

# 学習データセットとテストデータセットに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

# インスタンスを作成して学習
iristree = DecisionTreeClassifier(max_depth=3)
iristree.fit(X_train, y_train)
```

```python
# 予測（決定木分析）
y_pred1 = iristree.predict(X_test)
y_pred1
```

```python
# 予測結果の正解率（決定木分析）
print (iristree.score(X_test, y_test))
```

# ▶︎ ランダムフォレスト

```python
from sklearn.ensemble import RandomForestClassifier

# インスタンスを作成して学習
irisforest = RandomForestClassifier(n_estimators=100, random_state=123)
irisforest.fit(X_train, y_train)
```

```python
# 予測（ランダムフォレスト）
y_pred2 = irisforest.predict(X_test)
y_pred2
```

```python
# 予測結果の正解率（ランダムフォレスト）
print (irisforest.score(X_test, y_test))
```
