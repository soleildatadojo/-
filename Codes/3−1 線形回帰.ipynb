```
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```
# Bostonデータセットを読み込む
boston = load_boston()

print(type(boston))
```

```
print(boston.keys())
```

```
print(boston['data'])
```

```
print(boston['target'])
```

```
print(boston['feature_names'])
```

```
print(boston['DESCR'])
```

```
print(boston['filename'])
```

```
X = boston.data
y = boston.target

print(type(X))
print(type(y))
print(X.shape)
print(y.shape)
```

```
import pandas as pd
df_X = pd.DataFrame(X, columns = boston.feature_names)
df_X
```

```
# 学習データセットとテストデータセットに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```
df_Xtrain = pd.DataFrame(X_train, columns = boston.feature_names)
df_Xtrain
```

```
# 線形回帰をインスタンス化
bos = LinearRegression()
# 学習
bos.fit(X_train, y_train)
```

```
y_pred = bos.predict(X_test)
```

```
print(y_pred)
```

```
print(y_pred.shape)
print(X_test.shape)
```

```
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,7.5))
ax.scatter(y_pred, y_test)
ax.plot((0, 50), (0, 50), linestyle='dashed', color='limegreen')
ax.set_xlabel('predicted value')
ax.set_ylabel('actual value')
plt.title('boston house value')
plt.show()
```
