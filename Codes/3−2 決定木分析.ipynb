```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
```

```
# Irisデータセットを読み込む
iris = load_iris()
```

```
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_target = pd.DataFrame(iris.target, columns=["花の種類"])
iris_all = pd.concat([iris_data,iris_target], axis=1)
iris_all
```

```
print(iris.data.shape)
print(iris.target.shape)
print(iris.target_names)
```

```
X = iris.data
y = iris.target
```

```
# アヤメデータセットを学習データとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
df_iristrain = pd.DataFrame(X_train, columns = iris.feature_names)
df_iristrain
```

```
from sklearn.tree import DecisionTreeClassifier
iris3 = DecisionTreeClassifier(max_depth=3)
iris3.fit(X_train, y_train)
```

```
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data
# dot形式のデータを抽出
dot_data = export_graphviz(iris3, filled=True,
                rounded=True,
                class_names=['Setosa',
                     'Versicolor',
                     'Virigica'],
                feature_names=['Sepal Length',
                               'Sepal Width',
                               'Petal Length',
                               'Petal Width'],
                out_file=None)
# 決定木のプロットを出力
graph = graph_from_dot_data(dot_data)
graph.write_png('iris3.png')
```

```
plt.scatter(X_train[y_train == 0][:, 0],X_train[y_train == 0][:, 1],label="setosa")
plt.scatter(X_train[y_train == 1][:, 0],X_train[y_train == 1][:, 1],label="versicolor")
plt.scatter(X_train[y_train == 2][:, 0],X_train[y_train == 2][:, 1],label="virginica")

plt.xlim(3,9)
plt.ylim(1,5)

plt.xlabel("Sepaｌ Length")
plt.ylabel("Sepal Width")
plt.title("Relation between length and width of Sepal")

plt.legend()
```

```
plt.scatter(X_train[y_train == 0][:, 2],X_train[y_train == 0][:, 3],label="setosa")
plt.scatter(X_train[y_train == 1][:, 2],X_train[y_train == 1][:, 3],label="versicolor")
plt.scatter(X_train[y_train == 2][:, 2],X_train[y_train == 2][:, 3],label="virginica")

plt.xlim(0,8)
plt.ylim(0,4)

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Relation between length and width of petal")

plt.legend()
```

```
y_pred = iris3.predict(X_test)
y_pred
```

```
df_iristest = pd.DataFrame(X_test, columns=iris.feature_names)
df_irispred = pd.DataFrame(y_pred, columns=['花の種類（予測結果）'])
df_irisanswer = pd.DataFrame(y_test, columns=['花の種類（正解）'])
df_answer = pd.concat([df_iristest, df_irispred, df_irisanswer], axis=1)
df_answer
```
